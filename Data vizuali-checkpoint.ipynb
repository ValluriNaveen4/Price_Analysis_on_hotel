# Treating the outliers

```python
Q1, Q2,Q3= (totaloyo["Original_price_in(₹)"].quantile(q = [0.25,0.5,0.75]))
```

```python
print(Q1,Q2,Q3)
```

```python
IQR = Q3-Q1
rightwhisker = Q3 + 1.5*IQR
leftwhisker = Q1-1.5*IQR

print(IQR, rightwhisker, leftwhisker)
```

```python
totaloyo[(totaloyo["Original_price_in(₹)"] > rightwhisker) | (totaloyo["Original_price_in(₹)"] < leftwhisker)]
```

```python
totaloyo[(totaloyo["Original_price_in(₹)"]> rightwhisker) | (totaloyo["Original_price_in(₹)"] < leftwhisker)].index.tolist()
```
